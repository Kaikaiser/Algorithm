## 问题 4：

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9



idea:
    1.遍历一次数组：从前往后遍历，记录最低点作为买入点，然后寻找之后的最高点作为卖出点。
    2.记录最小值和最大利润：在遍历过程中，维护一个当前的最小值（初始设为第一天的值），以及最大利润（初始设为负无穷大）。每次遍历时，比较当前价格与最小值的差值是否为最大利润，如果是则更新最大利润以及对应的卖出日期。
    3.返回结果：遍历结束后，返回最大利润和对应的买入、卖出日期。

````python
# add your code
# Kadane’s Algorithm 算法
def max_profit(profits):
    n = len(profits)
    if n == 0:
        return 0, -1, -1  # 没有数据

    # 变量初始化
    max_profit = 0 # 最大收益
    min_accumulated = 0  # 当前最小累计收益
    min_index = -1  # 记录最小累计收益出现的天数
    accumulated_sum = 0  # 当前累计收益
    buy_day, sell_day = -1, -1  # 记录买入和卖出日期

    for i in range(n):
        accumulated_sum += profits[i]  # 计算累计收益

        # 计算当前天卖出的收益
        current_profit = accumulated_sum - min_accumulated
        if current_profit > max_profit:
            max_profit = current_profit
            buy_day = min_index + 1  # 买入天数
            sell_day = i  # 卖出天数

        # 更新最小累计收益
        if accumulated_sum < min_accumulated:
            min_accumulated = accumulated_sum
            min_index = i

    return max_profit, buy_day + 1, sell_day + 1  # 题目天数从 1 开始计数

# 示例测试
profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
profit, buy, sell = max_profit(profits)
print(f"最佳买入日: 第 {buy} 天, 最佳卖出日: 第 {sell} 天, 最大收益: {profit}")
````

#question 1:
    1.时间复杂度递归公式：T(n) = 3T(2n/3) + O(1),时间复杂度：O(n^{log_{3/2}3})，约等于O(n^2.7)
    2.该算法是一个递归实现的排序算法。它通过三次递归调用处理数组的前2/3和后2/3重叠部分，最终将整个数组排序。

#question 2:
否。该算法的时间复杂度为 O(n^2.7)远高于常见的高效排序算法（如归并排序、快速排序的 O(nlogn)）。因此，存在更高效的解决方案。

````python
#question 3:
# algorithm time complexity is :O(n^2.7)
import math

def do_something(A, p, r):
    n = r - p + 1
    if n == 2:
        if A[p] > A[r]:
            A[p], A[r] = A[r], A[p]
    elif n >= 3:
        m = math.ceil(2 * n / 3)
        do_something(A, p, p + m - 1)
        do_something(A, r - m + 1, r)
        do_something(A, p, p + m - 1)

# 测试
A = [3, 1, 2, 6, 9, 8, 10, 5]
do_something(A, 0, len(A)-1)
print(A)  # 输出: [1, 2, 3, 5, 6, 8, 9, 10]
````

````python
#question 3:
# algorithm time complexity is :O(nlogn)
#快速排序
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

# 示例
A = [3, 1, 2, 6, 9, 8, 10, 5]
merge_sort(A)
print(A)  # 输出: [1, 2, 3, 5, 6, 8, 9, 10]
````

e.g.:
---
1. nums=[3,2,3], 返回3
2. nums=[2,2,1,1,1,2,2], 返回2


#question1:
idea：初始化一个候选元素和一个计数器。遍历数组，如果当前元素与候选元素相同，则计数器加1；否则，计数器减1。当计数器为0时，更新候选元素为当前元素，并将计数器重置为1。最终的候选元素即为多数元素。

````python
# algorithm time complexity：O(n), space complexity:O(1)
def findMajorityElement(ans):
    result = 0
    count = 0
    for num in ans:
        if count == 0:
            result = num
            count = 1
        elif num == result:
            count =count + 1 
        else:
            count=count-1
    return result

# 测试
print(findMajorityElement([3,2,3])) # 输出 5
print(findMajorityElement([2, 2, 1, 1, 1, 2, 2]))  # 输出 2
````

1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现

#question 1；
idea：题目给定数组两端满足 A[0] > A[1] 和 A[n-1] < A[n]，说明数组左侧是下降趋势，右侧是上升趋势，中间必然存在至少一个局部最小值。利用二分法缩小搜索范围。

````python
# your algorithm time complexity is :O(log n)
def findLocalMin(arr):
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        # 检查中间点是否为局部最小
        if (mid == 0 or arr[mid] < arr[mid-1]) and (mid == len(arr)-1 or arr[mid] < arr[mid+1]):
            return arr[mid]
        # 向右半部分
        elif mid < len(arr)-1 and arr[mid] > arr[mid+1]:
            left = mid + 1
        # 向左半部分
        else:
            right = mid - 1
    return None 

# 测试
A = [9, 3, 7, 2, 1, 4, 5, 3]
print(findLocalMin(A))  # 输出 1
````

````python
#question 2:参考deepseek
#time complexity is :O((n+k)logn)

class MinHeap:
    def __init__(self):
        self.heap = []
    
    def parent(self, i):
        return (i-1)//2
    
    def left_child(self, i):
        return 2*i + 1
    
    def right_child(self, i):
        return 2*i + 2
    
    # 插入新元素并上浮调整
    def insert(self, val):
        self.heap.append(val)
        i = len(self.heap) - 1
        while i > 0 and self.heap[i] < self.heap[self.parent(i)]:
            self.heap[i], self.heap[self.parent(i)] = self.heap[self.parent(i)], self.heap[i]
            i = self.parent(i)
    
    # 提取堆顶并下沉调整
    def extract_min(self):
        if not self.heap:
            return None
        min_val = self.heap[0]
        self.heap[0] = self.heap[-1]
        self.heap.pop()
        self.heapify(0)
        return min_val
    
    # 递归下沉
    def heapify(self, i):
        left = self.left_child(i)
        right = self.right_child(i)
        smallest = i
        if left < len(self.heap) and self.heap[left] < self.heap[smallest]:
            smallest = left
        if right < len(self.heap) and self.heap[right] < self.heap[smallest]:
            smallest = right
        if smallest != i:
            self.heap[i], self.heap[smallest] = self.heap[smallest], self.heap[i]
            self.heapify(smallest)

# 主函数：使用堆找前k小
def find_k_smallest_manual_heap(arr, k):
    heap = MinHeap()
    for num in arr:
        heap.insert(num)  # 逐个插入建堆
    return [heap.extract_min() for _ in range(k)]

#测试
A = [9, 3, 7, 2, 1, 4, 5, 3]
print(find_k_smallest_manual_heap(A, 3))  # 输出 [1,2,3]
````

````python
#question 3: 参考deepseek
#time complexity is :O(nlogk)
class MaxHeap:
    def __init__(self, capacity):
        self.heap = []
        self.capacity = capacity
    
    def parent(self, i):
        return (i-1)//2
    
    def left_child(self, i):
        return 2*i + 1
    
    def right_child(self, i):
        return 2*i + 2
    
    # 插入新元素并上浮调整
    def insert(self, val):
        if len(self.heap) < self.capacity:
            self.heap.append(val)
            i = len(self.heap) - 1
            while i > 0 and self.heap[i] > self.heap[self.parent(i)]:
                self.heap[i], self.heap[self.parent(i)] = self.heap[self.parent(i)], self.heap[i]
                i = self.parent(i)
        else:
            if val < self.heap[0]:
                self.heap[0] = val
                self.heapify(0)
    
    # 下沉调整堆
    def heapify(self, i):
        left = self.left_child(i)
        right = self.right_child(i)
        largest = i
        if left < len(self.heap) and self.heap[left] > self.heap[largest]:
            largest = left
        if right < len(self.heap) and self.heap[right] > self.heap[largest]:
            largest = right
        if largest != i:
            self.heap[i], self.heap[largest] = self.heap[largest], self.heap[i]
            self.heapify(largest)

    # 获取堆中的元素（排序后）
    def get_sorted(self):
        return sorted(self.heap)

# 主函数：处理数据流
def find_k_smallest_external_manual(data_stream, k):
    max_heap = MaxHeap(k)
    for num in data_stream:
        max_heap.insert(num)
    return max_heap.get_sorted()

# 测试（模拟海量数据流）
data_stream = iter([5,4,3,2,6,1,88,33,22,107])
print(find_k_smallest_external_manual(data_stream, 3))  # 输出 [1,2,3]
````

````python
#question 1:参考deepseek
#algorithm time complexity is : O(nlogn)
import random
def partition(arr, low, high, pivot_idx):
    #以指定索引的值为pivot进行分割，返回分割后的中间索引
    pivot = arr[pivot_idx]
    arr[pivot_idx], arr[high] = arr[high], arr[pivot_idx]  # 将pivot移到末尾
    i = low
    for j in range(low, high):
        if arr[j] <= pivot:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
    arr[i], arr[high] = arr[high], arr[i]
    return i

def quick_select(arr, low, high, k):
    #手动实现线性时间选择算法（模拟BB函数）
    if low == high:
        return low
    pivot_idx = random.randint(low, high)  # 随机选择pivot（此处需优化为确定性选择）
    pivot_idx = partition(arr, low, high, pivot_idx)
    if k == pivot_idx:
        return k
    elif k < pivot_idx:
        return quick_select(arr, low, pivot_idx - 1, k)
    else:
        return quick_select(arr, pivot_idx + 1, high, k)

def modified_quicksort(arr, low, high):
    #改进的快速排序，每次选择中位数作为pivot
    if low < high:
        n = high - low + 1
        median_rank = (n // 2)  # 中位数的排名
        median_idx = quick_select(arr, low, high, low + median_rank)
        partition_idx = partition(arr, low, high, median_idx)
        modified_quicksort(arr, low, partition_idx - 1)
        modified_quicksort(arr, partition_idx + 1, high)

# 测试
arr = [13, 3, 7, 9, 11, 1, 15, 2, 8, 10, 12, 16, 14, 5]
modified_quicksort(arr, 0, len(arr)-1)
print("Sorted Array:", arr)
````

````python
#question 2:参考deepseek
#algorithm time complexity is : O(n + klogk)= O(n)
def find_k_smallest(arr, k):
    if k == 0:
        return []
    n = len(arr)
    # 找到第k小的元素索引
    k_idx = quick_select(arr.copy(), 0, n-1, k-1)  # 注意索引从0开始
    x = arr[k_idx]
    # 收集所有小于x的元素和等于x的元素
    less, equal = [], []
    for num in arr:
        if num < x:
            less.append(num)
        elif num == x:
            equal.append(num)
    # 合并结果并排序
    result = less + equal[:k - len(less)]
    result.sort()  # 排序k个元素，时间复杂度 O(k log k)
    return result[:k]

# 测试示例
arr = [13, 3, 7, 9, 11, 1, 15, 2, 8, 10, 12, 16, 14, 5]
print("前4小元素:", find_k_smallest(arr, 4))  # 输出 [1, 2, 3, 5]
````

````python
#question 3:参考deepseek
#algorithm time complexity is :O(nlogk)
def split_into_k_sections(arr, k):
    if k == 1:
        return [arr]
    n = len(arr)
    mid_rank = n // 2
    # 模拟BB函数：找到中位数索引（此处简化为排序后选择）
    sorted_part = sorted(arr)
    mid_val = sorted_part[mid_rank]
    # 分割数组为左（<=mid_val）和右（>mid_val）
    left = [x for x in arr if x <= mid_val]
    right = [x for x in arr if x > mid_val]
    # 递归处理左右部分
    return split_into_k_sections(left, k//2) + split_into_k_sections(right, k//2)

# 测试示例
arr = [1, 3, 5, 7, 9, 11, 13, 15, 2, 4, 6, 8, 10, 12, 16, 14]
k = 4
sections = split_into_k_sections(arr, k)
print("划分为{}个部分:".format(k), sections)
````

## 问题 10：

给定一个包含m个**字符串**的数组A, 其中不同的字符串可能有不同的字符数, 但数组中所有字符串的字符总数为n。设计一个算法在 $O(n)$ 时间内对字符串进行排序, 分析算法设计方案, 计算其时间复杂度, 并基于python编程实现该算法。请注意, 假设字符串只包含"a","b",...,"z", 



举例1：数组A=["a", "da", "bde", "ab", "bc", "abdc", "cdba"], 排序后的数组应该为：['a', 'ab', 'abdc', 'bc', 'bde', 'cdba', 'da']



举例2：数组A=['ab', 'a', 'b', 'abc', 'ba', 'c'], 排序后的数组应该为：

['a', 'ab', 'abc', 'b', 'ba', 'c']



举例3：数组A=['aef', 'yzr', 'wr', 'ab', 'bhjc', 'lkabdc', 'pwcdba'],  排序后的数组应该为：['ab', 'aef', 'bhjc', 'lkabdc', 'pwcdba', 'wr', 'yzr']



note：

-  两个字符之间的比较可以考虑比较他们对应的ASCII码值；
- python中可以使用`ord("a")`返回字符 “a”对应的ASCII值

idea:遍历整个数组，对每个字符串进行比大小操作，然后进行排序。

````python
#algorithm time complexity is :O(n~3)
def bigger(a,b):
    length=min(len(a),len(b))
    for i in range(0,length):
        if ord(a[i])>ord(b[i]):
            return b, a
        elif ord(a[i])<ord(b[i]):
            return a, b
    if len(a)>len(b):
        return b, a
    elif len(a)<=len(b):
        return a, b
            
def sortString(arr):
    for i in range(0,len(arr)):
        for j in range(0,len(arr)-i-1):
            arr[j],arr[j+1]=bigger(arr[j],arr[j+1])
    return arr

#测试
A=['aef', 'yzr', 'wr', 'w', 'ab', 'bhjc', 'lkabdc', 'pwcdba']
print(sortString(A))  #测试结果['ab', 'aef', 'bhjc', 'lkabdc', 'pwcdba', 'wr', 'yzr']
````
